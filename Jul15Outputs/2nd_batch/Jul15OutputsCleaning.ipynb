```python
import pandas as pd
import numpy as np
```

```python
errors = pd.read_csv("errors.csv")
errors
```

```python
import os
clubs = errors["CLUB"].value_counts().index
for club in clubs: 
    club_errors = errors[errors["CLUB"] == club]
    club_errors = club_errors["SHORTCODE"]
    file = open("errors/" + club + ".txt", "w")
    file.write("shortcodes_"+ club + " = [")
    for error in club_errors:
        file.write("\"" + error + "\",")
    file.write("]")
    file.close()
```
