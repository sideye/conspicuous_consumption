```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# To add new clubs, only change this cell by adding defining the club's table below, then adding it to the clubs dictionary.
rudas = pd.read_csv("Jul15Outputs/clean_outputs/data_rudas_posts.csv")
prysm = pd.read_csv("Jul15Outputs/clean_outputs/data_prysmCH_posts.csv")
cavesduroy = pd.read_csv("Jul15Outputs/clean_outputs/data_cavesduroyST_posts.csv")
grandBoston = pd.read_csv("Jul15Outputs/clean_outputs/data_grandBoston_posts.csv")
grandSF = pd.read_csv("Jul15Outputs/clean_outputs/data_grandSF_posts.csv")
oxford = pd.read_csv("Jul15Outputs/clean_outputs/data_oxfordsc_sd_posts.csv")
templeSF = pd.read_csv("Jul15Outputs/clean_outputs/data_templeSF_posts.csv")
wetrepublic = pd.read_csv("Jul15Outputs/clean_outputs/data_wetrepublic_posts.csv")
clubs = {"rudas": rudas, "prysm": prysm, "cavesduroy": cavesduroy, "grandBoston": grandBoston, "grandSF": grandSF, "oxford": oxford, "templeSF": templeSF, "wetrepublic": wetrepublic}
```

```python
# Number of posts:
bars = []
counts = []
for club in clubs:
    count = len(clubs[club].index)
    print(club, count)
    bars.append(club)
    counts.append(count)
plt.bar(bars, counts)
plt.title("Number of Posts")
plt.xlabel("Club")
plt.ylabel("# of Posts")
```

```python
# Number of Likes
bars = []
likes = []
data = []
for club in clubs:
    like_num = np.mean(clubs[club]["LIKES"])
    data.append(clubs[club]["LIKES"])
    bars.append(club)
    likes.append(like_num)
fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].bar(bars, likes)
axes[0].set_title("Average Number of Likes Per Post")
axes[0].set_xlabel("Club")
axes[0].set_ylabel("# of Likes")
axes[1].boxplot(data, labels = bars)
axes[1].set_xlabel("Club")
axes[1].set_ylabel("# of Likes")
axes[1].set_title("Average Number of Likes Per Post")
plt.show()
```

```python
# Average comments per post:
bars = []
comments = []
data = []
for club in clubs:
    comment_num = np.mean(clubs[club]["NUM_COMMENTS"])
    bars.append(club)
    comments.append(comment_num)
    data.append(clubs[club]["NUM_COMMENTS"])
fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].bar(bars, comments)
axes[0].set_title("Average Number of Comments Per Post")
axes[0].set_xlabel("Club")
axes[0].set_ylabel("# of Comments")
axes[1].boxplot(data, labels = bars)
axes[1].set_xlabel("Club")
axes[1].set_ylabel("# of Likes")
axes[1].set_title("Average Number of Comments Per Post")
plt.show()
```

```python
# Posts By Day of Week
fig, axes = plt.subplots(4,2, figsize=(15,15))
order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
for counter, club in enumerate(clubs):
    x = counter % 4
    y = counter // 4
    ax = axes[x,y]
    data = [clubs[club]["WEEKDAY"].value_counts().loc[i] for i in order]
    ax.bar(order, data)
    ax.set_title("Number of Posts by Day for " + club)
plt.figure(figsize=(8,8))
```

```python
# Most Common Mentions for Each Club. Seems a little buggy during data collection.
import re
for club in clubs:
    users = {}
    mentions_col = clubs[club]["MENTIONS"]
    for post in mentions_col:
        if isinstance(post, str):
            mentions = re.findall("(.+?)\s", post)
            for mention in mentions:
                if mention in users:
                    users[mention] += 1
                else:
                    users[mention] = 1
    results = sorted(users.items(), key=lambda x:x[1], reverse=True)
    print(club, "most often mentions: ", [(result[0],result[1]) for result in results[:5]])
```

```python
# Likes by Gender
fig, axes = plt.subplots(4,2, figsize=(8,15))
order = ["male", "female"]
for counter, club in enumerate(clubs):
    x = counter % 4
    y = counter // 4
    ax = axes[x,y]
    data = [clubs[club]["GENDER"].value_counts().loc[i] for i in order]
    ax.bar(order, data)
    ax.set_title("Likes by Gender for " + club)
plt.figure(figsize=(8,8))
```
