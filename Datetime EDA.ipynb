```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
```

# Loading and Cleaning

```python
# Using datas from the outputs2 folder
rudas = pd.read_csv('outputs2/data_rudas_posts.csv')
temple = pd.read_csv('outputs2/data_templeSF_posts.csv')
```

```python
# Note: there is a weird spacing error from writing the file, so that every value has a space before it.
# I have fixed it, but the results are still the original files.
rudas_times = rudas[' DATE']
temple_times = temple[' DATE']
rudas_times = [string[1:] for string in rudas_times]
temple_times = [string[1:] for string in temple_times]
```

```python
# Convert into datetime objects
rudas_times = [datetime.datetime.strptime(date,'%Y-%m-%d %H:%M:%S') for date in rudas_times]
temple_times = [datetime.datetime.strptime(date,'%Y-%m-%d %H:%M:%S') for date in temple_times]
```

# Date time EDA

## Day of Week

```python
weekdays = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
rudas_day = pd.Series([weekdays[date.weekday()] for date in rudas_times])
temple_day = pd.Series([weekdays[date.weekday()] for date in temple_times])
```

```python
# Rudas
weekday_series = rudas_day.value_counts()
plt.bar(weekday_series.index, weekday_series)
plt.title("Frequency of Posts by Weekday for Rudas")
```

```python
# Temple
weekday_series = temple_day.value_counts()
plt.bar(weekday_series.index, weekday_series)
plt.title("Frequency of Posts by Weekday for Temple")
```

## Time of Day

```python
rudas_hours = pd.Series([date.hour for date in rudas_times])
temple_hours = pd.Series([date.hour for date in temple_times])
```

```python
# Rudas
sns.distplot([date.hour + date.minute/60 for date in rudas_times])
plt.xticks(range(0,25,2))
plt.title("Frequency of Posts by Hour for Rudas")
plt.show()
```

### Temple

```python
# Temple
sns.distplot([date.hour + date.minute/60 for date in temple_times])
plt.xticks(range(0,25,2))
plt.title("Frequency of Posts by Hour for Temple")
plt.show()
```
